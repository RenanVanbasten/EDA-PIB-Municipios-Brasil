```python
import pandas as pd;
```

```python
# Utilizando o Pandas para ler um arquivo CSV e o argumento nrows para limitar o número de linhas lidas.

pd.read_csv('../data/pib_municipios_br_2020.csv', nrows=5)
```

```python
# Atribuindo o DataFrame para uma variável.

df = pd.read_csv('../data/pib_municipios_br_2020.csv')
```

```python
# Alterando os nomes das colunas, removendo caracteres de nova linha.

df.columns = df.columns.str.replace('\n', ' ')
```

```python
df.head(5)
```

```python
# Alterando o nome da segunda coluna, removendo caracteres desnecessários.

df.columns = [col.replace("ocu- pada", "ocupada") for col in df.columns]
df.head(5)
```

```python
# Mudando valores das colunes para que possam ser convertidos para numéricos.

df['Posição ocupada pelos 100 maiores municípios'] = df['Posição ocupada pelos 100 maiores municípios'].str.replace('º','')
df['Produto Interno Bruto a preços correntes (1 000 R$)'] = df['Produto Interno Bruto a preços correntes (1 000 R$)'].str.replace(' ','')
df['Participação (%)'] = df['Participação (%)'].str.replace(',' ,'.')
df['Participação acumulada (%)'] = df['Participação acumulada (%)'].str.replace(',' ,'.')
df.head(5)
```

```python
# Alterando os valores das colunas para o tipo numérico adequado.

df['Participação (%)'] = df['Participação (%)'].astype(float)
df['Participação acumulada (%)'] = df['Participação acumulada (%)'].astype(float)
df['Produto Interno Bruto a preços correntes (1 000 R$)'] = df['Produto Interno Bruto a preços correntes (1 000 R$)'].astype(int)
df['Posição ocupada pelos 100 maiores municípios'] = df['Posição ocupada pelos 100 maiores municípios'].astype(int)
```

```python
df.info()
```

```python
df['Municípios e respectivas Unidades da Federação'].head(5)
```

```python
# Criando duas novas colunas, uma para o nome da cidade e outra para o estado, a partir da coluna 'Municípios e respectivas Unidades da Federação'.

df['Cidades'] = df['Municípios e respectivas Unidades da Federação'].str.split('(').str[0]  
df['Estados'] = df['Municípios e respectivas Unidades da Federação'].str.split('(').str[1].str.replace(')', '')
df
```

```python
# Criando uma coluna para a região do estado.

regioes = {
    'Norte': ['AC', 'AP', 'AM', 'PA', 'RO', 'RR', 'TO'],
    'Nordeste': ['AL', 'BA', 'CE', 'MA', 'PB', 'PE', 'PI', 'RN', 'SE'],
    'Centro-Oeste': ['DF', 'GO', 'MT', 'MS'],
    'Sudeste': ['ES', 'MG', 'RJ', 'SP'],
    'Sul': ['PR', 'RS', 'SC']
}

estado_para_regiao = {estado: regiao for regiao, estados in regioes.items() for estado in estados}

df['Regioes'] = df['Estados'].map(estado_para_regiao)
```

```python
df.columns
```

```python
# Ordenando as colunas do DataFrame para melhor visualização.

colunas_ordenadas = ['Cidades', 'Estados', 'Regioes', 
                     'Posição ocupada pelos 100 maiores municípios', 
                     'Produto Interno Bruto a preços correntes (1 000 R$)', 
                     'Participação (%)', 
                     'Participação acumulada (%)']

df = df[colunas_ordenadas]
```

```python
# Verificando o número total de municípios por Região.

df.Regioes.value_counts()
```

```python
# Verificando o número total de municípios por Estado.

df.Estados.value_counts()
```

```python
# Calculando o Pib total por Estado e Região.

pib_por_estado = df.groupby('Estados')['Produto Interno Bruto a preços correntes (1 000 R$)'].sum()
pib_por_regiao = df.groupby('Regioes')['Produto Interno Bruto a preços correntes (1 000 R$)'].sum()
```

```python
pib_total_por_regiao_coluna = df.groupby('Regioes')['Produto Interno Bruto a preços correntes (1 000 R$)'].transform('sum')
pib_total_por_estado_coluna = df.groupby('Estados')['Produto Interno Bruto a preços correntes (1 000 R$)'].transform('sum')
```

```python
# Calculando e criando a coluna de contribuição do Estado na sua respectiva Região.

df['Contribuicao_Estado_na_Regiao (%)'] = ((pib_total_por_estado_coluna / pib_total_por_regiao_coluna) * 100).round(2)
df.head(30)
```

```python
# Visualizando a contribuição do Estado na sua respectiva Região.

contribuicao_por_regiao_estado = df.groupby(['Regioes', 'Estados'])['Contribuicao_Estado_na_Regiao (%)'].mean().round(2)
contribuicao_por_regiao_estado
```
